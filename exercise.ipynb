```python
import numpy as np
print(np.__version__)
np.__config__.show()
```

```python
Z = np.zeros(10)
Z[4] = 1
print(Z)
```

```python
# 创建一个向量，范围是 2 - 22
Z = np.arange(21) + 2
print(Z)
```

```python
# 创建一个三维矩阵，数据为 0 - 8
Z = np.arange(9)
Z = Z.reshape(3, 3)
print(Z)
```

```python
# 找出向量中所有非零的索引
Z = np.arange(21)
print(Z.nonzero())
```

```python
# 声明一个单位矩阵
Z = np.eye(10)
print(Z)
```

```python
# 声明一个 5x5 矩阵，对角线上下四个元素的值分别为 1、2、3、4
Z = np.zeros((5, 5))
V = np.arange(4) + 1
Z += np.diag(V, 1)
Z += np.diag(V, -1)
print(Z)
```

```python
# 声明一个 10 * 10 * 10 矩阵，填入随机值
Z = np.random.random((10, 10, 10))
print(Z)
```
